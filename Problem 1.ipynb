```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
num_steps = 500
num_walks = 10
step_size = 1

plt.figure(dpi=300,figsize=(8,4))
for i in range(num_walks):
    walk = np.zeros(num_steps)
    steps = np.random.normal(loc=0, scale=step_size, size=num_steps)
    walk = np.cumsum(steps)
    plt.plot(walk)

plt.title("Multiple Brownian Motion Processes")
plt.xlabel("Step")
plt.ylabel("Location")
plt.show()
```
